```python

#1.  c
#2. Expanise each pixel by 10*10
def grey2rgb(original_path,org_veg_path,filename):
    # read pic and get shape size
    original_veg_path=os.path.join(original_path, filename)
    img = imread(original_veg_path)
    #print(original_veg_path)
    w=img.shape[0]
    h=img.shape[1]
    if(h==250 or w==250):
        print(filename)
        os.remove(os.path.join(original_path, filename))
        os.remove(os.path.join(org_img_path, filename.replace('veg1','image')))
    else:
        print(filename)
        img_array=np.zeros((w*10,h*10,3), dtype=np.uint8)

        for i in range(0,h):
            for j in range(0,w):               
                cls_color=color_dict[img[i][j][0]]# get RGB color of classification
                pixels = np.full((10,10,3),fill_value=cls_color,dtype=np.uint8) # get 10*10 pixels 
    #             pixels = np.zeros((10,10,3),dtype=np.uint8) # get 10*10 pixels 
    #             pixels[:,:]=cls_color
                img_array[i*10:i*10+10,j*10:j*10+10]=pixels

        img = Image.fromarray(img_array, 'RGB')
        img.save(os.path.join(org_veg_path, filename))
```

```python
import glob
files = glob.glob(org_veg_path+'*') 
for f in files:
    os.remove(f)
```
